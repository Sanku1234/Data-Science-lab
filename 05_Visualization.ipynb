```python
df = pd.read_csv("financialportfolio.csv")
```

```python
df.head(20)
```

```python
#convert 'Data' to datatime

df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.dropna(subset=['Date'])
```

```python
df = df.sort_values(by='Date').reset_index(drop=True)
```

```python
plt.figure(figsize=(14, 7))
for asset in df['Asset'].unique():
    asset_data = df[df['Asset'] == asset]
    plt.plot(asset_data['Date'], asset_data['Price'], label=asset)

    plt.title("Time series of ASSET PRICES")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#filter date for year only the 2023
plt_2023 = df[df['Date'].dt.year == 2023]
```

```python
plt.figure(figsize=(14, 7))

# Get list of unique assets
assets = plt_2023['Asset'].unique()

# Create a color palette (one color for each asset)
colors = plt.cm.tab10.colors

# Loop through each asset
for i, asset in enumerate(assets):
    asset_data = plt_2023[plt_2023['Asset'] == asset]

    # Resample to monthly average
    monthly_avg = asset_data.set_index('Date')['Price'].resample('ME').mean()

    # Format index to month names (e.g., Jan, Feb, ...)
    monthly_avg.index = monthly_avg.index.strftime('%b')

    # Plot bar chart for each asset separately (side by side)
    plt.bar(
        monthly_avg.index,
        monthly_avg.values,
        color=colors[i % len(colors)],
        alpha=0.6,
        label=asset,
        width=0.15
    )

plt.title("Monthly Average Price by Asset in 2023", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Average Price", fontsize=12)
plt.legend(title="Assets", fontsize=10, title_fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()




```

```python
df = df.sort_values(by="Date").reset_index(drop=True)
```

```python
df['YearMonth'] = df['Date'].dt.to_period('M')
```

```python
# Create a pivot table to calculate the average price of each Asset by Year-Month

pivot_df = df.pivot_table(
    index='YearMonth',
    columns='Asset',
    values='Price',
    aggfunc='mean'
)

plt.figure(figsize=(12, 6))

# Plot a heatmap showing average asset prices over time
sns.heatmap(pivot_df, annot=True, cmap='coolwarm', fmt=".1f")
plt.title("Heatmap of Asset Prices Over Time", fontsize=14)
plt.xlabel("Asset", fontsize=12)
plt.ylabel("Year-Month", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a boxplot showing the distribution of 'Price' for each 'Asset'

sns.boxplot(x='Asset', y='Price', data=df)
plt.title("Boxplot of Asset Prices", fontsize=16)
plt.xlabel("Asset", fontsize=14)
plt.ylabel("Price", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Bar Chat(Average Price per Assets)

avg_prices = df.groupby('Asset')['Price'].mean().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_prices.index, y=avg_prices, palette='viridis')
plt.title("Average Asset Prices", fontsize=16)
plt.xlabel("Assets")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
pivot_area = df.pivot_table(
    index='Date',
    columns='Asset',
    values='Price',
    aggfunc='mean'
).fillna(0)

plt.figure(figsize=(14,7))
pivot_area.plot.area(figsize=(14, 7), alpha=0.7)
plt.title("Portfolio composition Asset Prices Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Average Price", fontsize=12)
plt.legend(title="Assets", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.dates as mdates

#pivot_area: monthly averages for smoother area chat
pivot_area = df.pivot_table(
    index=pd.Grouper(key="Date", freq="ME"),
    columns="Asset",
    values="Price",
    aggfunc="mean"
).fillna(0)

#plot refined stacked area chart
plt.figure(figsize=(14, 7))
pivot_area.plot.area(alpha=0.8, cmap="tab20", linewidth=1)

#title and labels
plt.title("Monthly Average Asset Prices", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Average Price", fontsize=12)

#format x-axis to show Year+Month
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))
plt.xticks(rotation=45)

#legend
plt.legend(title="Assets", bbox_to_anchor=(1.02, 1), loc='upper left')

#grid and layout
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```
